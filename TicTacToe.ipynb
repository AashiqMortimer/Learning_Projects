```python
from IPython.display import clear_output

def display_board(board):
    clear_output()
    
    print("|-----------|")
    print('|   |   |   |')
    print('| '+board[1]+' | '+board[2]+' | '+board[3]+' |')
    print('|   |   |   |')
    print('|---|---|---|')
    print('|   |   |   |')
    print('| '+board[4]+' | '+board[5]+' | '+board[6]+' |')
    print('|   |   |   |')
    print('|---|---|---|')
    print('|   |   |   |')
    print('| '+board[7]+' | '+board[8]+' | '+board[9]+' |')
    print('|   |   |   |')
    print("|-----------|")
```

```python
def player_input():
    marker = 'wrong'
    icons = ['X', 'O']
    
    while marker not in icons:
        marker = input('Please select your preferred icon, X or O. ').upper()
        
        if marker not in icons:
            print("Sorry. That was an invalid selection. Please pick between X or O.")
            
    return marker
```

```python
def place_marker(board, marker, position):
    
    board[position] = marker
    
    return board
```

```python
def win_check(board, mark):
    if board[1] is mark and board[2] is mark and board[3] is mark:
        return True
    elif board[4] is mark and board[5] is mark and board[6] is mark:
        return True
    elif board[7] is mark and board[8] is mark and board[9] is mark:
        return True
    elif board[1] is mark and board[5] is mark and board[9] is mark:
        return True
    elif board[3] is mark and board[5] is mark and board[7] is mark:
        return True
    elif board[1] is mark and board[4] is mark and board[7] is mark:
        return True
    elif board[2] is mark and board[5] is mark and board[8] is mark:
        return True
    elif board[3] is mark and board[6] is mark and board[9] is mark:
        return True
    else:
        return False
```

```python
import random

def choose_first():
    num = random.randint(1,2)
    if num == 1:
        return 'Player One'
    else:
        return 'Player Two'
```

```python
def space_check(board, position):
    icons = ['X', 'O']
    
    return board[position] in icons
```

```python
def full_board_check(board):
    icons = ['X','O']
    
    return board[1] in icons and board[2] in icons and board[3] in icons and board[4] in icons and board[5] in icons and board[6] in icons and board[7] in icons and board[8] in icons and board[9] in icons
```

```python
def player_choice(board):
    position = 0
    valid = [1,2,3,4,5,6,7,8,9]
    
    while position not in valid or space_check(board, position) == True:
        
        position = int(input('Please pick a number between 1-9 to choose your next position. '))
        
        if position not in valid:
            print('Sorry, you have selected an incorrect number. Please choose between 1 and 9.')
        
        elif space_check(board, position) == True:
            print('Sorry, the space you have selected has already been taken. Please choose another position.')
    
    return position
```

```python
def replay():
    gameon = 'wrong'
    valid = ['Y','N']
    
    while gameon not in valid:
        gameon = input('Would you like to play again (Y/N)? ').upper()
        if gameon not in valid:
            print('Sorry, I could not understand that. Please choose between Y or N.')
            
    if gameon == 'Y':
        return True
    elif gameon == 'N':
        return False
```

```python
game_on = 'game'
board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
marker_1 = ''
marker_2 = ''

while True:
    #Randomise who will be selecting their icon.
    print('Welcome to Tic Tac Toe!')
    first = choose_first()
    
    #Icon selection by player number.
    if first == 'Player One':
        print('***Player One***')
        marker_1 = player_input()
        
        if marker_1 == 'X':
            marker_2 = 'O'
            game_on = 'game'
        elif marker_1 == 'O':
            marker_2 = 'X'
            game_on = 'game'
    elif first == 'Player Two':
        print ('***Player Two***')
        marker_2 = player_input()
        if marker_2 == 'X':
            marker_1 = 'O'
            game_on = 'game'
        elif marker_2 == 'O':
            marker_1 = 'X'
            game_on = 'game'
        
    #Player 1 Turn
    while game_on == 'game':
        print("Player 1: It's your turn.")
        position = player_choice(board)
        
        board = place_marker(board, marker_1, position)
        
        display_board(board)
        
        #Need to check whether board is full or game won before continuing to Player 2
        
        if win_check(board, marker_1) == True:
            print('Player 1 wins!')
            restart = replay()
            if restart:
                print('Restarting game')
                board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
                continue
            elif not restart:
                game_on = 'end'
                print('Thank you for playing.')
                break
                
    
        elif full_board_check(board) == True:
            print("It's a draw!")
            restart = replay()
            if restart:
                print('Restarting game')
                board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
                continue
            elif not restart:
                game_on = 'end'
                print('Thank you for playing.')
                break
            
        else:
            print("Player 2: It's your turn.")
        
        # Player2's turn.
        
        position = player_choice(board)
        
        board = place_marker(board, marker_2, position)
        
        display_board(board)
        
        if win_check(board, marker_2) == True:
            print('Player 2 wins!')
            restart = replay()
            if restart:
                print('Restarting game')
                board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
                continue
            elif not restart:
                game_on = 'end'
                print('Thank you for playing.')
                break
        
        elif full_board_check(board) == True:
            print("It's a draw!")
            restart = replay()
            if restart:
                print('Restarting game')
                board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
                continue
            elif not restart:
                game_on = 'end'
                print('Thank you for playing.')
                break
        
    if game_on == 'end':
        break
        #End of game. 
```
